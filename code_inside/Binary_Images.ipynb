### Khai báo các thư viện 

```python
from IPython.display import Image
import matplotlib.pyplot as plt
from matplotlib.colors import NoNorm

import cv2 as cv
import numpy as np
print(cv.__version__)
```

Erosion


```python
img = cv.imread('images/image-j.png',0)
kernel = np.ones((5,5),np.uint8)
erosion = cv.erode(img,kernel,iterations = 1)
plt.figure(figsize=(12,12))
plt.subplot(121)
plt.title('Original')
plt.imshow(img,cmap='gray')
plt.subplot(122)
plt.title('After erode')
plt.imshow(erosion,cmap='gray')
```

Dilation

```python
img = cv.imread('images/circles.png',0)
kernel = np.ones((3,3),np.uint8)
dilation = cv.dilate(img,kernel,iterations = 1)
plt.figure(figsize=(12,12))
plt.subplot(121)
plt.title('Original')
plt.imshow(img,cmap='gray')
plt.subplot(122)
plt.title('After dilation')
plt.imshow(dilation,cmap='gray')

kernel = np.ones((7,7),np.uint8)
dilation = cv.dilate(img,kernel,iterations = 1)
plt.figure(figsize=(12,12))
plt.subplot(121)
plt.title('Original')
plt.imshow(img,cmap='gray')
plt.subplot(122)
plt.title('After dilation')
plt.imshow(dilation,cmap='gray')
```

Opening


```python
img = cv.imread('images/image-j.png',0)
kernel = np.ones((5,5),np.uint8)
opening = cv.morphologyEx(img, cv.MORPH_OPEN, kernel)
plt.figure(figsize=(12,12))
plt.subplot(121)
plt.title('Original')
plt.imshow(img,cmap='gray')
plt.subplot(122)
plt.title('After dilation')
plt.imshow(opening,cmap='gray')
```

Closing


```python
img = cv.imread('images/image-j.png',0)
kernel = np.ones((5,5),np.uint8)
closing = cv.morphologyEx(img, cv.MORPH_CLOSE, kernel)
plt.figure(figsize=(12,12))
plt.subplot(121)
plt.title('Original')
plt.imshow(img,cmap='gray')
plt.subplot(122)
plt.title('After dilation')
plt.imshow(closing,cmap='gray')
```

```python
img=cv.imread('images/fence.jpg',0)
blur = cv.GaussianBlur(img,(5,5),0)
ret, otsu = cv.threshold(blur,0,255,cv.THRESH_BINARY+cv.THRESH_OTSU)
kernelSizes = (151, 151)
kernel = cv.getStructuringElement(cv.MORPH_CROSS, kernelSizes);
erosion = cv.erode(otsu,kernel,iterations = 1)
plt.figure(figsize=(12,12))
plt.subplot(131)
plt.title('Original')
plt.imshow(img,cmap='gray')
plt.subplot(132)
plt.title('Otsu')
plt.imshow(otsu,cmap='gray', vmin=0, vmax=255)
plt.subplot(133)
plt.title('after erosion')
plt.imshow(erosion,cmap='gray', vmin=0, vmax=255)

contours,hierarchy = cv.findContours(erosion, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)
for i, c in enumerate(contours):
    area = cv.contourArea(c)
    print (area)
    x,y,w,h = cv.boundingRect(c)
    print (x, y, w, h)



```
