```
#include libraries
import cv2 as cv
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
import glob 
#Load an image in the directory
img=cv.imread('images/Bk2017.jpg')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.figure(figsize = (10,10))
plt.imshow(img)
plt.show()
#Convert color to gray image
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
plt.figure(figsize = (10,10))
plt.imshow(gray, cmap='gray', vmin=0, vmax=255)
plt.show()
#compute color histogram
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.savefig("images/color_histogram_Bk2017.jpg") 
#compute histogram of a grayscale image
histr = cv.calcHist([gray],[0],None,[256],[0,256])
plt.plot(histr)
plt.show()
#compute histogram and cdf of a grayscale image using matplotlib
hist,bins = np.histogram(gray.flatten(),256,[0,256])
cdf = hist.cumsum()
cdf_normalized = cdf * hist.max()/ cdf.max()
plt.plot(cdf_normalized, color = 'b')
plt.hist(gray.flatten(),256,[0,256], color = 'r')
plt.xlim([0,256])
plt.legend(('cdf','histogram'), loc = 'upper left')
#Save image
plt.savefig("images/histogram_Bk2017.jpg")      
```
